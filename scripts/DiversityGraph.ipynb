```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [
    # 1gram
    [321, 332, 350, 355, 356],
    # 2gram
    [319, 332, 334, 345, 345],
    # 3gram
    [322, 327, 331, 348, 349],
    # 4gram
    [322, 329, 334, 336, 336],
    # 5gram
    [327, 341, 343, 344, 352],
    # 6gram
    [319, 330, 336, 349, 351]
]
```

```python
x_labels = [10, 11, 12, 13,14]
```

```python
plt.figure(figsize=(10, 6))
for i, productions in enumerate(data):
    plt.plot(x_labels, productions, label=f"{i+1} gram")
plt.xticks(x_labels, [f"2^{x}" for x in x_labels])
plt.scatter(13.5, 427, color='red', marker='x', label='Chibench')
plt.scatter(11.5, 137, color='blue', marker='x', label='VlogHammer')
plt.scatter(14, 179, color='green', marker='x', label='Verismith')
plt.scatter(14, 456, color='orange', marker='x', label='Verilog Grammar')
plt.xlabel('Number of programs (2^n)')
plt.ylabel('Total Productions')
plt.title('Total Productions vs. Number of programs')
plt.legend()
plt.grid(True)
plt.savefig('productions.png', format='png', dpi=300)
plt.show()
```

```python
data = [
    [164, 180, 194, 197, 201],  # 1gram tokens
    [160, 174, 180, 187, 194],  # 2gram tokens
    [157, 168, 176, 185, 192],  # 3gram tokens
    [165, 176, 191, 195, 195],  # 4gram tokens
    [162, 174, 185, 193, 197],  # 5gram tokens
    [159, 172, 184, 195, 199]   # 6gram tokens
]
```

```python
plt.figure(figsize=(10, 6))
for i, tokens in enumerate(data):
    plt.plot(x_labels, tokens, label=f"{i+1} gram")
plt.scatter(13.5, 246, color='red', marker='x', label='Chibench')
plt.scatter(11.5, 58, color='blue', marker='x', label='VlogHammer')
plt.scatter(14, 57, color='green', marker='x', label='Verismith')
plt.scatter(14, 299, color='orange', marker='x', label='Verilog Grammar')
plt.xticks(x_labels, [f"2^{x}" for x in x_labels])
plt.xlabel('Number of programs(2^n)')
plt.ylabel('Total Tokens')
plt.title('Total Tokens vs. Number of programs')
plt.legend()
plt.grid(True)
plt.savefig('tokens.png', format='png', dpi=300)
plt.show()
```
