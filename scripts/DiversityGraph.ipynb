```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [
    # 1-gram productions
    [331, 350, 376, 394, 406],
    # 2-gram productions
    [306, 338, 342, 366, 390],
    # 3-gram productions
    [313, 332, 340, 369, 391],
    # 4-gram productions
    [352, 354, 361, 379, 389],
    # 5-gram productions
    [354, 359, 368, 391, 400],
    # 6-gram productions
    [320, 323, 350, 385, 401]
]
```

```python
x_labels = [10, 11, 12, 13,14]
```

```python
fig, ax = plt.subplots(figsize=(10, 6), layout='constrained')

# Plot data using ax.plot
for i, productions in enumerate(data):
    ax.plot(x_labels, productions, label=f"{i+1} gram")

# Customize ticks
ax.set_xticks(x_labels)
ax.set_xticklabels([f"2^{x}" for x in x_labels], fontsize=13)
ax.tick_params(axis='y', labelsize=13)

# Add scatter points
ax.scatter(13.5, 427, color='red', marker='x', label='Chibench')
ax.scatter(11.5, 137, color='blue', marker='x', label='VlogHammer')
ax.scatter(14, 179, color='green', marker='x', label='Verismith')
ax.scatter(14, 456, color='orange', marker='x', label='Verilog Grammar')

# Set labels and title
ax.set_xlabel('Number of programs (2^n)', fontsize=15)
ax.set_ylabel('Total Productions', fontsize=15)
ax.set_title('Total Productions vs. Number of programs', fontsize=17)

# Add legend, grid, and save the figure
fig.legend(loc='outside center right', prop={'size': 13})
ax.grid(True)
fig.savefig('productions.png', format='png', dpi=300)

# Show plot
plt.show()
```
