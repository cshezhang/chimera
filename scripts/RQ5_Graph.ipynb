```python
import matplotlib.pyplot as plt

# X-axis labels
x_labels = ['v9', 'v10', 'v11', 'v12', 'v13', 'v14', 'v15', 'v16']
x = list(range(len(x_labels)))

# Data for 1-gram and 6-gram
tokens_1gram = [10076, 12727, 12397, 40752, 101851, 41650, 71963, 92426]
tokens_6gram = [22999, 23762, 20445, 43794, 45092, 50267, 84984, 79480]

percentage_valid_1gram = [77.9, 88.9, 86.9, 28.1, 61.1, 27.5, 44.6, 37.9]
percentage_valid_6gram = [58.9, 68.9, 68.2, 28.3, 28.8, 32.1, 51.6, 49.0]

avg_size_valid_1gram = [12.93, 14.32, 14.27, 145.02, 166.7, 151.45, 161.35, 157.45]
avg_size_valid_6gram = [39.05, 34.49, 29.98, 154.75, 156.57, 156.6, 164.7, 162.2]

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot Number of Tokens on the primary y-axis
ax1.plot(x, tokens_1gram, marker='o', linestyle='-', color='b', label='Tokens (1-gram)')
ax1.plot(x, tokens_6gram, marker='s', linestyle='-', color='r', label='Tokens (6-gram)')
ax1.set_xlabel('Versions (v9 to v16)')
ax1.set_ylabel('Number of Tokens', color='black')
ax1.tick_params(axis='y')
ax1.grid(True)

# Create a second y-axis for Percentage of Valid Programs
ax2 = ax1.twinx()
ax2.plot(x, percentage_valid_1gram, marker='o', linestyle='--', color='g', label='Percentage Valid (1-gram)')
ax2.plot(x, percentage_valid_6gram, marker='s', linestyle='--', color='orange', label='Percentage Valid (6-gram)')
ax2.set_ylabel('Percentage of Valid Programs (%)', color='black')
ax2.tick_params(axis='y')

# Create a third y-axis for Average Size of Valid Programs (sharing the same x-axis)
ax3 = ax1.twinx()
ax3.spines['right'].set_position(("outward", 60))  # Offset the third y-axis
ax3.plot(x, avg_size_valid_1gram, marker='o', linestyle='-.', color='purple', label='Avg Size (1-gram)')
ax3.plot(x, avg_size_valid_6gram, marker='s', linestyle='-.', color='brown', label='Avg Size (6-gram)')
ax3.set_ylabel('Average Size of Valid Programs', color='black')
ax3.tick_params(axis='y')

# Combine legends from all y-axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines3, labels3 = ax3.get_legend_handles_labels()
ax1.legend(lines1 + lines2 + lines3, labels1 + labels2 + labels3, loc='upper left')

# Set x-axis labels
plt.xticks(x, x_labels)
plt.title('Comparison of 1-gram and 6-gram Data Across Multiple Metrics')
plt.tight_layout()
plt.show()
```
