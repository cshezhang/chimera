```python
import matplotlib.pyplot as plt

# Data for formatted and no formatted times
gram = [1, 2, 3, 4, 5, 6]
time_formatted = [8.5, 8.9, 16.347, 16.429, 12.883, 16.631]
time_no_formatting = [3.961, 4.368, 6.685, 7.855, 8.34, 8.608]

# Plotting the data
plt.figure(figsize=(10, 4))

# Line for formatted data
plt.plot(gram, time_formatted, marker='o', label='Formatting', linewidth=2, linestyle='-')

# Line for no formatting data
plt.plot(gram, time_no_formatting, marker='o', label='No Formatting', linewidth=2, linestyle='-')

# Adding titles and labels
plt.xlabel("n-gram", fontsize=15)
plt.ylabel("Time (s)", fontsize=15)

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

# Adding grid and legend
plt.grid(True)
plt.legend(fontsize=13)

# Display the plot
plt.show()
```
